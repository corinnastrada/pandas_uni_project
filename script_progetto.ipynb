7. For each book in goodbooks, compute the list of its authors. Assuming that the number of reviews with a text (column work_text_reviews_count) is split equally among all authors, find for each authors the total number of reviews with a text. We will call this quantity the shared number of reviews with a text.

```python
for i, row in goodbooks.iterrows():
    goodbooks.at[i,'authors'] = row["authors"].split(',')
    num_of_authors = len(row["authors"].split(','))
    goodbooks.at[i,'work_text_reviews_count'] = row["work_text_reviews_count"] / num_of_authors
```

```python
goodbooks_exploded = goodbooks.explode('authors')
```

```python
goodbooks_exploded.head()
```

```python
goodbooks_exploded.columns
```

```python
authors_review= goodbooks_exploded.groupby('authors', as_index=False)['work_text_reviews_count'].sum()
```

```python
authors_review
```

```python
authors_review.rename(columns={'work_text_reviews_count': 'shared_number_of_reviews_with_a_text'}, inplace=True)
```

```python
authors_review.sort_values(by='authors', ascending=True)
```
